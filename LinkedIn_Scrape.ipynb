```python
SCROLL_PAUSE_TIME = 5

# Get scroll height
last_height = browser.execute_script("return document.body.scrollHeight")

for i in range(3):
    # Scroll down to bottom
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = browser.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
src = browser.page_source
soup = BeautifulSoup(src, 'lxml')
```

```python
name_div = soup.find_all('div', {'class': 'search-result__wrapper'})
for row in name_div:          # Print all occurrences
    print(row)
```

```python
df = pd.DataFrame()
for row in name_div:
    name_loc = row.find('span', {'class': 'name actor-name'})
    print("Name: ", name_loc.get_text())
    role = row.find('span', {'dir': 'ltr'})
    print("Present role: ",role.get_text())
    loc = row.find('p', {'class': 'subline-level-2 t-12 t-black--light t-normal search-result__truncate'})
    print("Present Location: ",loc.get_text())
    skill=row.find('p',{'class':'mt2 t-12 t-black--light t-normal search-result__snippets-black'})
    print(skill.get_text())
    location = loc.get_text()
    location = location.strip("\n")
    skills = skill.get_text()
    skills = skills.strip("\n")
    df = df.append({'Name':name_loc.get_text(),'Present role':role.get_text(),'Present Location':location,'Skills':skills},ignore_index=True)  
```

```python
print(df)
df.to_csv("linkds1.csv")
```
