```python
import requests, time, random
from bs4 import BeautifulSoup
from selenium import webdriver
import pandas as pd
```

```python
browser = webdriver.Chrome('chromedriver.exe')
```

```python
link = 'https://angel.co/search?q=Deep+Learning&type=people'
# link = 'https://www.linkedin.com/in/simran-gandhi-024593128/'
# link = 'https://www.linkedin.com/in/smriti-handa-0636045/'
# link = 'https://www.linkedin.com/in/charan-nadendla-858a17120/'
# link = 'https://www.linkedin.com/in/mayank-agrawal-1a5883148/'
browser.get(link)
```

```python
SCROLL_PAUSE_TIME = 5

# Get scroll height
last_height = browser.execute_script("return document.body.scrollHeight")

for i in range(3):
    # Scroll down to bottom
    browser.execute_script("window.scrollTo(0, document.body.scrollHeight);")

    # Wait to load page
    time.sleep(SCROLL_PAUSE_TIME)

    # Calculate new scroll height and compare with last scroll height
    new_height = browser.execute_script("return document.body.scrollHeight")
    if new_height == last_height:
        break
    last_height = new_height
```

```python
src = browser.page_source
soup = BeautifulSoup(src, 'lxml')
```

```python
name_div = soup.find_all('div', {'class': 'section g-module_section ds30 frt71 _a'})
for row in name_div:          # Print all occurrences
    print(row)
```

```python
df = pd.DataFrame() 
for row in name_div:
    name_loc = row.find('div', {'class': 'title'})
    name=name_loc.get_text()
    name=name.split('-')
    print("Name: ", name[0])
    if(len(name)>1):
        print("Role: ", name[1])
    skills=row.find('div',{'class':'type'})
    skills=skills.get_text()
    skills=skills.split('·')
    print("Skills: ", skills[0])
    if(len(skills)>1):
        print("Location: ", skills[1])
    skills[1] = skills[1].strip("\n")
    skills[0] = skills[0].strip("\n")
    '''role = row.find('span', {'dir': 'ltr'})
    print("Present role: ",role.get_text())
    loc = row.find('p', {'class': 'subline-level-2 t-12 t-black--light t-normal search-result__truncate'})
    print("Present Location: ",loc.get_text())
    skills=row.find('p',{'class':'mt2 t-12 t-black--light t-normal search-result__snippets-black'})
    print(skills.get_text())'''
    df = df.append({"Name":name[0],"Present Role":name[1],"Present Location":skills[1],"Skills":skills[0]},ignore_index=True)
print(df)
```
